# Random Matrix examples

following https://random-matrix-learning.github.io/pdf/part1.pdf

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Gaussian Orthogonal Ensemble

```python
n = 10000
A=np.random.randn(n,n)
GOE=(A+A.T)/np.sqrt(2*n)
eig=np.linalg.eigvals(GOE)


fig, ax = plt.subplots(figsize=(10, 5))

ax.hist(
    eig,
    bins=80,
    color= "blue",
    edgecolor="white",
    linewidth=0.4,
    alpha=0.85,
)

plt.show()
```

# Wishart ensemble - Marchenko-Pastur distribution

```python
n = 10000
r=1/2 #for example
X=np.random.randn(int(n* r),n)
W=np.dot(X,X.T)/n
eig=np.linalg.eigvals(W)

fig, ax = plt.subplots(figsize=(10, 5))

ax.hist(
    eig,
    bins=80,
    color= "blue",
    edgecolor="white",
    linewidth=0.4,
    alpha=0.85,
)

plt.show()
```
